```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,1000)
y = np.sin(x)
ax.plot(x,y)
```

**Matplotlib Multiple Plots**

```python
plt.style.use('classic')
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y + y2 - 0.5
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
```

```python
fig = plt.figure(figsize=(4,3)) # figsize(weidth,height )
ax = plt.axes()
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

**Matplotlib Colors And Styles**

```python
plt.plot(x, np.sin(x-0), color='blue')  
plt.plot(x, np.sin(x- 1), color='g')  # g represent green color
plt.plot(x, np.sin(x - 2), color=(1.0,0.2,0.3)) # tuple for rgb colors 

# curve style
plt.plot(x, x + 0, linestyle='solid')  # solid curve
plt.plot(x, x + 1, linestyle='dashed')  # dashed curve
plt.plot(x, x + 2, linestyle='dashdot')  # dash and dot 
plt.plot(x, x + 3, linestyle='dotted')  # purely dot
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes()
ax.plot(x, y1, color= 'blue', linestyle='solid')
ax.plot(x, y2, color= 'g',linestyle='dashed')
ax.plot(x, y3, color= (1,0,0), linestyle='dashdot')
```

**Matlotlib Colors And Styles Shortcuts**

```python
plt.plot(x, x + 4, linestyle='-') # solid  
plt.plot(x, x + 5, linestyle='--') # dashed  
plt.plot(x, x + 6, linestyle='-.') # dashdot  
plt.plot(x, x + 7, linestyle=':'); # dotted  


plt.plot(x, x + 0,'-g') # solid green  
plt.plot(x, x + 1,'--c') # dashed cyan  
plt.plot(x, x + 2,'-.k') # dashdot black  
plt.plot(x, x + 3,':r') #dotted red
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.axes()

# ax.plot(x, y1, color= 'blue', linestyle='solid')
ax.plot(x, y1,'-b', linewidth=2)

# ax.plot(x, y2, color= 'g',linestyle='dashed')
ax.plot(x, y2, '--g', linewidth=1.5)

# ax.plot(x, y3, color= (1,0,0), linestyle='dashdot')
ax.plot(x, y3, ':r')
```

**Labels and legends**

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-b', label='cos(x)')

plt.legend()
plt.title('Sine-Cos Curve')
```

```python
plt.figure(figsize=(6,4),dpi = 100)
plt.plot(x,y1, '--r', linewidth = 0.8, label ='sin(x)')
plt.plot(x,y2, '-.b', linewidth = 0.8, label ='cos(x)')
plt.plot(x,y3, ':g', linewidth = 0.8, label ='sin(x)+cos(x)-0.5')
plt.xlim(-2,12)
plt.ylim(-3,1.5)
# plt.axis='equal'
plt.legend(loc='lower left')
plt.title('Sin-Cos_Curve')
plt.xlabel('X values')
plt.ylabel('Y values')
```
