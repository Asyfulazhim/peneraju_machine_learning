```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('Medals.csv')

totalmedals = np.array(df['Total'])
totalmedalrange = [10, 20, 30, 40]  # Ranges for the histogram


def plot_histogram(data, bins, title, x_label, y_label):
    plt.figure(figsize=(10, 6))
    plt.hist(data, bins=bins, alpha=1)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    # plt.grid(axis='y', , alpha=0.7)
    #plt.xticks(bins)
    #plt.show()


if __name__ == "__main__":
    title = 'Total Medals Histogram'
    x_label = 'Total Medals Range'
    y_label = 'Count'
    
    plot_histogram(totalmedals, totalmedalrange, title, x_label, y_label)
```
